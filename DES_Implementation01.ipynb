```python
# Sample DES implementation in Python (simplified)

# Initial Permutation table
IP = [58, 50, 42, 34, 26, 18, 10, 2,
      60, 52, 44, 36, 28, 20, 12, 4,
      62, 54, 46, 38, 30, 22, 14, 6,
      64, 56, 48, 40, 32, 24, 16, 8,
      57, 49, 41, 33, 25, 17, 9, 1,
      59, 51, 43, 35, 27, 19, 11, 3,
      61, 53, 45, 37, 29, 21, 13, 5,
      63, 55, 47, 39, 31, 23, 15, 7]

# Final Permutation table
FP = [40, 8, 48, 16, 56, 24, 64, 32,
      39, 7, 47, 15, 55, 23, 63, 31,
      38, 6, 46, 14, 54, 22, 62, 30,
      37, 5, 45, 13, 53, 21, 61, 29,
      36, 4, 44, 12, 52, 20, 60, 28,
      35, 3, 43, 11, 51, 19, 59, 27,
      34, 2, 42, 10, 50, 18, 58, 26,
      33, 1, 41, 9, 49, 17, 57, 25]

# Permute function according to a table
def permute(block, table):
    return ''.join(block[i - 1] for i in table)

# XOR two bit strings
def xor(bit_string1, bit_string2):
    return ''.join('1' if bit_string1[i] != bit_string2[i] else '0' for i in range(len(bit_string1)))

# Simplified function to represent a round (not full DES, just for understanding)
def simple_round(left, right, key):
    # Use XOR of right half and key (simplified Feistel structure)
    new_right = xor(right, key)
    new_left = right  # Swap halves
    return new_left, new_right

# Encrypt function (simplified, only 1 round)
def des_encrypt(plain_text, key):
    # Convert plain_text to binary
    binary_text = ''.join(format(ord(char), '08b') for char in plain_text)

    # Apply initial permutation
    permuted_text = permute(binary_text, IP)

    # Split into two halves
    left, right = permuted_text[:32], permuted_text[32:]

    # Perform one round (simplified)
    left, right = simple_round(left, right, key)

    # Combine left and right
    combined_text = left + right

    # Apply final permutation
    cipher_text = permute(combined_text, FP)

    return cipher_text

# Decrypt function (simplified, reverse process)
def des_decrypt(cipher_text, key):
    # Apply initial permutation (since it's symmetric)
    permuted_text = permute(cipher_text, IP)

    # Split into two halves
    left, right = permuted_text[:32], permuted_text[32:]

    # Perform one round in reverse (simplified)
    right, left = simple_round(left, right, key)

    # Combine left and right
    combined_text = left + right

    # Apply final permutation
    plain_text_binary = permute(combined_text, FP)

    # Convert binary back to characters
    plain_text = ''.join(chr(int(plain_text_binary[i:i + 8], 2)) for i in range(0, len(plain_text_binary), 8))

    return plain_text

# Example usage
key = '10101010101010101010101010101010'  # Simplified binary key (32 bits)
plain_text = "HiDES"

# Encrypt the message
cipher_text = des_encrypt(plain_text, key)
print(f"Encrypted (binary): {cipher_text}")

# Decrypt the message
decrypted_text = des_decrypt(cipher_text, key)
print(f"Decrypted Text: {decrypted_text}")
```

Initial Permutation (IP): use a table (IP) to permute (rearrange) the input binary text.

Key: A simple binary key is provided ('101010...'), and this is XORed with the right half of the input in the encryption round.

Feistel Round: This is the core part of DES, where the right half is XORed with the key, and the halves are swapped.

Final Permutation (FP): After processing, use a final permutation table (FP) to further rearrange the bits and produce the cipher text.

Decryption: Decryption is just reversing the encryption process using the same key and operations in reverse.
